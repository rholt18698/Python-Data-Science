```python
import numpy as np
from numpy.polynomial.polynomial import Polynomial
import matplotlib.pyplot as plt

# Create a polynomial from coefficients -0.8, 2.3, 0.5, 1
# https://en.wikipedia.org/wiki/Polynomial 
p = Polynomial([-0.8, 2.3, 0.5, 1])
p_der = p.deriv() # return a derivative of this polynomial.

fig, ax = plt.subplots()
X = np.linspace(-2, 3, 50, endpoint=True) # return evenly spaced numbers over a specified interval
F = p(X)
F_derivative = p_der(X)
ax.plot(X, F)
ax.plot(X, F_derivative)

ax.legend(['p', 'derivation of p'])
```

```python
p = Polynomial([-0.8, 2.3, 0.5, 1])
p_der = p.deriv()

fig, ax = plt.subplots()
X = np.linspace(-2, 3, 50, endpoint=True)
F = p(X)
F_derivative = p_der(X)
ax.plot(X, F, label="p")                           # THIS LINE CHANGED
ax.plot(X, F_derivative, label="derivation of p")  # THIS LINE CHANGED

ax.legend(loc='center')                        # THIS LINE CHANGED
```

```python
p = Polynomial([0, -12, 0, 1])
p_der = p.deriv()

fig, ax = plt.subplots()
X = np.arange(-5, 5, 0.1)
F = p(X)

F_derivative = p_der(X)
ax.grid() # adds grid lines

maximum = (-2, p(-2))
minimum = (2, p(2))
ax.annotate("local maximum", maximum) # (text to add, where text is placed)
ax.annotate("local minimum", minimum) # (text to add, where text is placed)

ax.plot(X, F, label="p")
ax.plot(X, F_derivative, label="derivation of p")

ax.legend(loc='best')
plt.show()
```

```python
p = Polynomial([0, -12, 0, 1])
p_der = p.deriv()

fig, ax = plt.subplots()
X = np.arange(-5, 5, 0.1)
F = p(X)

F_derivative = p_der(X)
ax.grid()

ax.annotate("local maximum", 
            xy=(-2, p(-2)),          # The point (x, y) to annotate
            xytext=(-1, p(-2)+35),   # The position (x, y) to place the text
            arrowprops=dict(facecolor='orange')) # draws an arrow between the positions xy and xytext
ax.annotate("local minimum", 
            xy=(2, p(2)), 
            xytext=(-2, p(2)-40),
            arrowprops=dict(facecolor='orange', shrink=0.05))
ax.annotate("inflection point",                 # shrink= fraction of total length to 
            xy=(0, p(0)),                       # shrink from both ends
            xytext=(-3, -30),
            arrowprops=dict(facecolor='blue', shrink=0.05)) 
                                                                  
ax.plot(X, F, label="p")
ax.plot(X, F_derivative, label="derivation of p")

ax.legend(loc='best')
plt.show()
```

```python
import matplotlib.ticker as ticker

X = np.linspace(-2 * np.pi, 2 * np.pi, 100)
F1 = np.sin(X)  # Trigonometric sine
F2 = 3 * np.sin(X)
fig, ax = plt.subplots()

positions = [np.pi / 2 * x for x in range(-4, 5)]
labels = [r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', 
                    r'$-\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', 
                    r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$']
ax.xaxis.set_major_locator(ticker.FixedLocator(positions))
                           # https://matplotlib.org/3.1.1/gallery/ticks_and_spines/tick-locators.html
ax.xaxis.set_major_formatter(ticker.FixedFormatter(labels))
                           # https://matplotlib.org/3.1.1/gallery/ticks_and_spines/tick-formatters.html

ax.plot(X, F1, label="$sin(x)$")
ax.plot(X, F2, label="$3 sin(x)$")
ax.legend(loc='lower left')
x = 3 * np.pi / 4

# Plot vertical line:
ax.plot([x, x],[-3, 3 * np.sin(x)], color ='blue', linewidth=2.5, linestyle="--")
# Print the blue dot:
ax.scatter([x,],[3 * np.sin(x),], 50, color ='blue')

text_x, text_y = (3.5, 2.2)
ax.annotate(r'$3\sin(\frac{3\pi}{4})=\frac{3}{\sqrt{2}}$',
            xy=(x, 3 * np.sin(x)),
            xytext=(text_x, text_y), 
            arrowprops=dict(facecolor='orange', shrink=0.05),
            fontsize=12)

plt.show()
```

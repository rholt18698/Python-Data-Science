```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

import pandas as pd
excel_file='Cerial Top.csv'
df= pd.read_csv(excel_file)
print(df)

fig, ax = plt.subplots()

Name = df['name']
Rating= df['rating']
bar_labels = ['red', 'blue']
bar_colors = ['tab:red','tab:blue']

ax.bar(Name, Rating) 
label=("bar_labels", 'color=bar_colors')
ax.set_ylabel('Rating')
ax.set_title("Top 15 Cereal's")
plt.xticks(rotation = 90)
ax.set_xlabel("Cereal", color="red")
ax.set_ylabel("Rating", color="green")



plt.show()
```
